#### Detect Outliers

```python
# IQR RANGE
for k, v in df[numerical_columns].items():
        q1 = v.quantile(0.25)
        q3 = v.quantile(0.75)
        inter_q = q3 - q1
        v_col = v[(v <= q1 - 1.5 * inter_q) | (v >= q3 + 1.5 * inter_q)]
        perc = np.shape(v_col)[0] * 100.0 / np.shape(df)[0]
        print("Column %s outliers = %.2f%%" % (k, perc))
```

#### Handling class imbalance

Synthetic Minority Oversampling Technique : It involves finding the K Nearest Neighbours of minority instance and selecting one randomly and producing a new datapoint b/w these two.

```python
from imblearn.over_sampling import SMOTE
oversampler = SMOTE(random_state = 0)
df1_minmax , y1 = oversampler.fit_resample(df1_minmax,y)
```

```python
sns.countplot(y1)
```

```python
## using df2 data set
df2_minmax , y2 = oversampler.fit_resample(df2_minmax,y)
```

#### VIF Factor

we create a new data frame which will include all the VIFs note that each variable has its own variance inflation factor as this measure is variable specific (not model specific) we do not include categorical values for mulitcollinearity as they do not provide much information as numerical ones do

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variables = df1_minmax.values
vif = pd.DataFrame()
# here we make use of the variance_inflation_factor, which will basically output the respective VIFs 
vif["VIF"] = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]
# Finally, I like to include names so it is easier to explore the result
vif["Features"] = df1_minmax.columns
```
